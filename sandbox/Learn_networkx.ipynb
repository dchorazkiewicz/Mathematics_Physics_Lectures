```python
import networkx as nx
import matplotlib.pyplot as plt
```

### Add nodes

```python
G=nx.Graph() # define a graph

o1 = G.add_node('o1', label="sdsd") # add a node
o2 = G.add_node('o2') # add a node
o3 = G.add_node('o3') # add a node
print(G.nodes(data=True))
```

```python
fig = plt.figure(figsize=(3, 3))
nx.draw(G, with_labels=True)
plt.show()
```

### Add edges

```python
G = nx.Graph()

G.add_edge('o1', 'o2', resistance="R1") # add an edge
G.add_edge('o2', 'o3', resistance="R2") # add an edge
print(G.edges(data=True))
```

```python
print(G['o1']['o2']) # access the edge data
```

```python
G.edges
```

```python
fig = plt.figure(figsize=(3, 3))
nx.draw(G, with_labels=True)
plt.show()
```

### Dense graph

```python
# drwa random dense graph with many nodes and edges
G = nx.dense_gnm_random_graph(10, 20)
# plot graph
nx.draw(G, with_labels=True)
plt.show()
```

```python
G.edges(data=True)
for u, v, d in G.edges(data=True):
    print(f"edge {u} to {v} has data {d}")
```

```python
G.nodes(data=True)
```

```python
for node in G.nodes():
    print(f"Node {node} has {G.degree(node)} edges")
```
