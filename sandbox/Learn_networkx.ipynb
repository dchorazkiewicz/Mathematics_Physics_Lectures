```python
import networkx as nx
import matplotlib.pyplot as plt
```

### Add nodes

```python
G=nx.Graph() # define a graph

o1 = G.add_node('o1', label="sdsd") # add a node
o2 = G.add_node('o2') # add a node
o3 = G.add_node('o3') # add a node
print(G.nodes(data=True))
```

```python
fig = plt.figure(figsize=(3, 3))
nx.draw(G, with_labels=True)
plt.show()
```

### Add edges

```python
G = nx.Graph()

G.add_edge('o1', 'o2', resistance="R1") # add an edge
G.add_edge('o2', 'o3', resistance="R2") # add an edge
print(G.edges(data=True))
```

```python
fig = plt.figure(figsize=(3, 3))
nx.draw(G, with_labels=True)
plt.show()
```
